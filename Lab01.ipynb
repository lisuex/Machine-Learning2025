```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

url = "https://github.com/SzymonNowakowski/Machine-Learning-2024/raw/master/PCA_three_clusters_data.csv"
data = pd.read_csv(url)
```

```python
X = data[['x', 'y', 'z']]
y = data['class']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("\nExplained variance ratio for each component:")
print(pca.explained_variance_ratio_)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA: Projection of Data onto 2 Principal Components")
plt.colorbar(label="Class")
plt.show()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
feature_pairs = [('x', 'y'), ('x', 'z'), ('y', 'z')]
for i, (f1, f2) in enumerate(feature_pairs):
    axes[i].scatter(data[f1], data[f2], c=y, cmap='viridis', alpha=0.7)
    axes[i].set_xlabel(f1)
    axes[i].set_ylabel(f2)
    axes[i].set_title(f"{f1} vs {f2}")
plt.suptitle("Original Feature Pairs")
plt.show()
```

```python
print("\nAnswers to the questions:")

# Primary question
print("\n1. Does PCA help separate clusters better than the original features?")
print("Yes, PCA helps because it finds the directions where the data varies the most. This often makes clusters clearer, especially when dealing with many features. The original features might not be as effective at separating them.")

# Secondary question
print("\n2. Why do we need to standardize data before using PCA?")
print("Because PCA is affected by the scale of features. If some features have much larger values than others, they will dominate the results. Standardizing puts all features on the same scale, so each one has a fair impact.")
```
